```python
import pandas as pd

df=pd.read_excel('titanic3.xls')
df.head()
```

```python
import numpy as np
df['female'] = np.where(df['sex']=='female',1,0)
df.head()
```

```python
def embark_pt(embarked):
    embarkpt = 0
    if embarked == 'C':
        embarkpt = 1
    elif embarked =='Q':
        embarkpt = 2
    elif embarked == 'S':
        embarkpt = 3
    return embarkpt
```

```python
df['embark_pt'] = df.embarked.apply(embark_pt)
```

```python
import statsmodels.formula.api as sm

result = sm.ols(formula = 'survived~pclass+female+embark_pt',data=df).fit()
result.summary()
```

```python
result = sm.ols(formula = 'survived~pclass+female+embark_pt-1',data=df).fit()
result.summary()
```

```python
result = sm.ols(formula = 'survived~pclass+female-1',data=df).fit()
result.summary()
```

```python
result = sm.ols(formula = 'survived~pclass-1',data=df).fit()
result.summary()
```

```python
result = sm.ols(formula = 'survived~pclass+embark_pt-1',data=df).fit()
result.summary()
```

```python
result = sm.ols(formula = 'survived~female+embark_pt-1',data=df).fit()
result.summary()
```
