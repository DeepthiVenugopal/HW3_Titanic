```python
survivor_wid_family = (train_df[(train_df['Survived']==1) & (train_df['Family']==1)]['PassengerId'].count())
passengers_wid_family =(train_df[train_df['Family']==1]['PassengerId'].count())
```

```python
survivor_wiot_family = (train_df[(train_df['Survived']==1) & (train_df['Family']==0)]['PassengerId'].count())
passengers_wiot_family =(train_df[train_df['Family']==0]['PassengerId'].count())
```

```python
#Survivor Ratio with Family
SRF = survivor_wid_family/passengers_wid_family

#Survivor Ratio without Family
SRNF = survivor_wiot_family/passengers_wiot_family
```

```python
bins = [0,15,30,50,80]
groupnames = ['Children','Young','Middle Aged','Old']
train_df['AllAgeGroup'] = pd.cut(train_df['Age'],bins,labels =groupnames)
train_df.head()
```

```python
child_df = train_df[train_df['Age'] <16]
```

```python
child_df.head()
```

```python
#Survival Rate
SR=(train_df[train_df['Survived']==1]['PassengerId'].count())/(train_df['PassengerId'].count())

#Survival Rate for children

CSR=(child_df[child_df['Survived']==1]['PassengerId'].count())/(child_df['PassengerId'].count())
```

```python
SR
```

```python
CSR
```

```python
child_df['Family']=np.where(child_df['Parch']>=1,1,0)
```

```python
childsurvivor_df=pd.DataFrame(child_df.groupby(['Survived','Family'])['PassengerId'].count())
childsurvivor_df.plot(kind='bar')
```

```python
child_df.groupby('Family')['PassengerId'].count()
```

```python
survivor_wiot_family = (child_df[(child_df['Survived']==1) & (child_df['Parch']==0)]['PassengerId'].count())
children_wiot_family =(child_df[child_df['Parch']==0]['PassengerId'].count())
```

```python
survivor_wit_family = (child_df[(child_df['Survived']==1) & (child_df['Parch']>0)]['PassengerId'].count())
children_wit_family =(child_df[child_df['Parch']>0]['PassengerId'].count())
```

```python
#Survivor Ratio with Family
CSRF = survivor_wit_family/children_wit_family

#Survivor Ratio without Family
CSRNF = survivor_wiot_family/children_wiot_family
```

```python
CSRF
```

```python
CSRNF
#Children travelling without family did better!
```

```python
plt.plot(SR,SR,'ro',label='SurvivalRate')
plt.plot(SRF,SRF,'r^',label = 'SRwFamily')
plt.plot(SRNF,SRNF,'rs',label='SRw/tFamily')
plt.plot(CSR,CSR,'go',label='ChildSurvivalRate')
plt.plot(CSRF,CSRF,'g^',label='CSRwFamily')
plt.plot(CSRNF,CSRNF,'gs',label='CSRw/tFamily')
plt.legend()
```
